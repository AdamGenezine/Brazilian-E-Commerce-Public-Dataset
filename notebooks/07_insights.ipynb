## Project Summary

Análise Exploratória de Dados do Brazilian E-Commerce Public Dataset da Olist,
cobrindo 100k pedidos de 2016 a 2018 em 9 tabelas relacionais.

# Insights

### Sales Analysis


- O negocio cresceu consistentemente de 2 pedidos em setembro/2016 para mais de
  7.000 pedidos mensais em 2018, evidenciando forte crescimento operacional.

- Novembro/2017 registrou o maior volume (7.451 pedidos), impulsionado pela Black Friday.

- bed_bath_table, health_beauty e computers_accessories sao as categorias com maior receita.

- A maioria dos clientes compra a vista — mediana de 1 parcela — mas produtos
  caros tendem a ser mais parcelados (correlacao de 0.28).

### Satisfaction Analysis


- Pedidos entregues no prazo tem nota media 62% maior que pedidos atrasados (4.14 vs 2.55).

- Nenhuma variavel financeira influencia a nota do cliente — frete caro nao derruba a nota.

- Estados do Norte e Nordeste concentram as piores avaliacoes.

- 75% dos clientes dao nota 5 — a maioria esta satisfeita, mas insatisfacoes sao intensas.

### Logistics Analysis

- 90.5% dos pedidos chegam antes do prazo estimado — a Olist trabalha com prazos conservadores.

- Estados do Norte tem tempo de entrega 3x maior que Sao Paulo (27.8 vs 8.3 dias).

- Janeiro concentra o maior numero de atrasos — reflexo da Black Friday e Natal.

- Entregas entre estados diferentes levam em media o dobro do tempo (14.6 vs 7.5 dias).

# Business Recommendations

1. Logistica regional: investir em centros de distribuicao no Norte e Nordeste
   para reduzir tempo de entrega e melhorar satisfacao nessas regioes.

2. Gestao de picos: reforcar a logistica em novembro e dezembro para evitar
   o acumulo de atrasos em janeiro.

3. Estrategia de categorias: priorizar estoque e logistica para bed_bath_table,
   health_beauty e computers_accessories — categorias de maior receita.

4. Prazo de entrega: manter a estrategia de prazos conservadores — 90% das
   entregas antes do prazo contribui diretamente para a satisfacao do cliente.

# Conclusions

O principal fator de satisfacao do cliente na Olist e o prazo de entrega —
nao o preco, nao o frete, mas simplesmente **receber o pedido no prazo**.

A desigualdade logistica regional e o maior desafio do negocio: enquanto SP
recebe pedidos em 8 dias, Roraima espera quase 28 dias pela mesma compra.

Resolver essa assimetria e o caminho mais direto para melhorar a experiencia
do cliente em escala nacional.
